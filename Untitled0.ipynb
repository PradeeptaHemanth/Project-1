```python
import speech_recognition as sr
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import pyjokes
!pip install SpeechRecognition pyttsx3 pywhatkit wikipedia

# Initialize recognizer and speech engine
listener = sr.Recognizer()
engine = pyttsx3.init()

# Set the voice property to a female voice
voices = engine.getProperty('voices')
engine.setProperty('voice', voices[1].id)

# Function to make the assistant speak
def talk(text):
    engine.say(text)
    engine.runAndWait()

# Function to listen for commands from the user
def take_command():
    command = ""  # Initialize 'command' to avoid UnboundLocalError
    try:
        with sr.Microphone() as source:
            print('Listening...')
            listener.adjust_for_ambient_noise(source)  # Adjust for ambient noise
            voice = listener.listen(source)  # Capture the user's voice
            command = listener.recognize_google(voice)  # Convert voice to text
            command = command.lower()  # Convert the command to lowercase
            if 'kaphia' in command:  # Process the command if it includes 'kaphia'
                command = command.replace('kaphia', '').strip()
                print(f"Command recognized: {command}")
    except sr.UnknownValueError:
        print("Sorry, I couldn't understand that.")
        command = ""  # Return an empty string if recognition fails
    except sr.RequestError:
        print("Sorry, there was an issue with the speech recognition service.")
        command = ""  # Return an empty string if there is a service error
    return command

# Function to run commands based on user input
def run_kaphia():
    command = take_command()  # Get the user's command
    print(f"Processing command: {command}")
    if command == "":  # If no command was recognized, ask the user to repeat
        talk("I didn't catch that. Please repeat.")
        return True  # Keep listening for commands

    if 'send a whatsapp message' in command:  # Send a WhatsApp message
        talk("To whom should I send the message? Please provide the phone number with country code.")
        phone_number = input("Enter the recipient's phone number (with country code, e.g., +1234567890): ").strip()
        talk("What message should I send?")
        message = input("Enter the message: ").strip()
        talk("At what time should I send it? Enter hours (24-hour format):")
        hour = int(input("Enter hour (24-hour format): "))
        talk("Now, enter the minutes:")
        minute = int(input("Enter minute: "))
        try:
            talk(f"Scheduling your message to {phone_number}.")
            pywhatkit.sendwhatmsg(phone_number, message, hour, minute)
            talk("Message scheduled successfully.")
        except Exception as e:
            talk(f"Sorry, I couldn't send the message. Error: {str(e)}")

    elif 'day' in command or 'date' in command:  # Tell the day or date if asked
        current_date = datetime.datetime.now()
        day = current_date.strftime('%A')  # Get the day of the week
        date = current_date.strftime('%d %B, %Y')  # Format date as 'day month, year'
        talk(f"Today is {day}, {date}.")
        print(f"Today is {day}, {date}.")

    elif 'play' in command:  # Play a song if the command contains 'play'
        song = command.replace('play', '').strip()
        talk(f'Playing {song}')
        pywhatkit.playonyt(song)

    elif 'time' in command:  # Tell the time if the command contains 'time'
        current_time = datetime.datetime.now().strftime('%I:%M %p')
        print(f"Current time: {current_time}")
        talk(f'Current time is {current_time}')

    elif 'who is' in command:  # Provide Wikipedia information if the command contains 'who is'
        person = command.replace('who is', '').strip()
        info = wikipedia.summary(person, 1)
        print(info)
        talk(info)

    elif 'what is' in command:  # Provide Wikipedia information if the command contains 'what is'
        detail = command.replace('what is', '').strip()
        fact = wikipedia.summary(detail, 1)
        print(fact)
        talk(fact)

    elif 'what are you doing' in command:  # Respond to 'what are you doing'
        talk('Just talking to my human friends')

    elif 'hello' in command:  # Respond to a hello greeting
        talk('Hello, what can I do for you?')

    elif 'tell me a joke' in command:  # Tell a joke if the command contains 'tell me a joke'
        joke = pyjokes.get_joke()
        print(f"Joke: {joke}")
        talk(joke)

    elif 'goodbye' in command:  # End the program if the command contains 'goodbye'
        talk('Goodbye')
        return False  # Exit the loop and end the program

    else:
        talk('Sorry, I didn’t understand that. Can you repeat?')
    return True  # Keep listening for new commands

# Run the assistant in a loop until 'goodbye' is said
while True:
    if not run_kaphia():
      break
```
